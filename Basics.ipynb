```python
odometer = autos['odometer_km'].unique()
avg_price_per_km_group = {}

for group in odometer:
    avg_price = autos.loc[autos['odometer_km'] == group, 'price'].mean()
    avg_price_per_km_group[group] = avg_price
    
for km_group in avg_price_per_km_group:
    print('{:.0f} km: \t${:.2f}'.format(km_group, avg_price_per_km_group[km_group]))
```

```python
avg_price_per_km_group_series = pd.Series(avg_price_per_km_group)
avg_price_per_km_group_series
```

```python
avg_price_per_km_group_series.round(2).sort_index()
```

### Observations 

* There is negative slope on the relationship between the price and the total mileage since the price drops when the mileage gets higher
* The lowest average price corresponds to the highest mileage

```python
brands = brand_model.unique()
damaged_avg_price = {}
no_damage_avg_price = {}

for brand in brands:
    avg_price = autos.loc[(brand_model == brand) & (autos['unrepaired_damage'] == 'yes'), 'price'].mean()
    damaged_avg_price[brand] = avg_price
    
for brand in brands:
    avg_price = autos.loc[(brand_model == brand) & (autos['unrepaired_damage'] == 'no'), 'price'].mean()
    no_damage_avg_price[brand] = avg_price
                                                       
damaged_avg_price_series = pd.Series(damaged_avg_price)
no_damage_avg_price_series = pd.Series(no_damage_avg_price)
df_damage_avg_price = pd.DataFrame(damaged_avg_price_series, columns = ['damaged'])
df_damage_avg_price['not_damaged'] = no_damage_avg_price_series
df_damage_avg_price.head()
```

```python
df_damage_avg_price['delta'] = df_damage_avg_price['not_damaged'] - df_damage_avg_price['damaged']
df_damage_avg_price
```

```python
df_damage_avg_price = df_damage_avg_price[df_damage_avg_price['delta'] != df_damage_avg_price['delta'].min()]
```

```python
df_damage_avg_price['delta'].describe()
```

```python
df_damage_avg_price[df_damage_avg_price['delta'] < 0].head()
```

### Observations

* There is a mean difference in price between a damaged and not damaged car of \$4,848.44 less.
* There are some scenerios where the difference in price is actually higher for damaged cars compared to not damaged car in the same brand/model category
